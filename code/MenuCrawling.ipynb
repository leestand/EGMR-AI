```python
def get_menu_info(store_id):
    url = 'https://m.place.naver.com/restaurant/' + store_id + '/review/visitor?entry=plt&reviewSort=recent'

    menu_info = dict()

    # Start crawling/scraping!
    try:
        driver = webdriver.Chrome()
        res = driver.get(url)
        driver.implicitly_wait(30)
        

        # 업종 추출
        time.sleep(1)
        try:
            menu1 = driver.find_element(By.XPATH, '//*[@id="_tag_filters"]/div/div[1]/div/div/div/div/span[2]/a/span[1]')
            menu2 = driver.find_element(By.XPATH, '//*[@id="_tag_filters"]/div/div[1]/div/div/div/div/span[3]/a/span[1]')
            menu3 = driver.find_element(By.XPATH, '//*[@id="_tag_filters"]/div/div[1]/div/div/div/div/span[4]/a/span[1]')
            menu_info['store_id'] = store_id
            menu_info['menu'] = [menu1.text, menu2.text, menu3.text]
            print("가게 메뉴:", menu_info['menu'])
        except Exception:
            menu_info['menu'] = []
            print("가게 메뉴: 없음")
        

        # 결과 반환
        return menu_info

    except Exception as e:
        print(e)
        return menu_info
```

```python
def get_multiple_menus(store_ids):
    all_menu_data = []  # 모든 가게 데이터를 저장할 리스트

    for store_id in store_ids:
        print(f"Scraping store: {store_id}")
        menu_dict = get_menu_info(store_id)
        
        # 가게 정보와 리뷰 데이터를 딕셔너리로 저장
        all_menu_data.append(menu_dict)

    return all_menu_data
```

```python
def get_category_info(store_id):
    url = 'https://m.place.naver.com/restaurant/' + store_id + '/review/visitor?entry=plt&reviewSort=recent'

    category_info = dict()

    # Start crawling/scraping!
    try:
        driver = webdriver.Chrome()
        res = driver.get(url)
        driver.implicitly_wait(30)
        

        # 업종 추출
        time.sleep(1)
        try:
            category = driver.find_element(By.XPATH, '//*[@id="_title"]/div/span[2]')
            category_info['store_id'] = store_id
            category_info['category'] = category.text
            print("가게 업종:", category_info['category'])
        except Exception:
            store_info['category'] = ''
            print("가게 업종: 없음")
        

        # 결과 반환
        return category_info

    except Exception as e:
        print(e)
        return category_info
```

```python
def multiple_category_info(store_ids):
    category_info_dict = []  # 모든 가게 데이터를 저장할 리스트

    for store_id in store_ids:
        print(f"Get store info: {store_id}")
        category_info = get_category_info(store_id)
        
        # 가게 정보를 딕셔너리로 저장
        category_info_dict.append(category_info)

    return category_info_dict
```
