```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

def crawl_review(driver, store_id, store_dict):
    review_list = []

    try:
        # íŽ˜ì´ì§€ ë‹¤ìš´ (ë¡œë”© ìœ ë„)
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
        time.sleep(2)

        # ë¦¬ë·° 30ë²ˆê¹Œì§€ 'ë”ë³´ê¸°' í´ë¦­
        try:
            for i in range(30):
                more_button = driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[6]/div[2]/div[3]/div[2]/div/a')
                more_button.click()
                time.sleep(2)
        except:
            print("ë”ë³´ê¸° í´ë¦­ ì¢…ë£Œ")
        else:
            print("30ë²ˆ í´ë¦­ ì™„ë£Œ")

        # HTML íŒŒì‹±
        time.sleep(2)
        html = driver.page_source
        bs = BeautifulSoup(html, 'lxml')

        # ë¦¬ë·° í•­ëª©ë“¤ ê°€ì ¸ì˜¤ê¸°
        reviews = bs.select('li.place_apply_pui.EjjAW')

        for r in reviews:
            # ë‹‰ë„¤ìž„
            nickname = r.select_one('.pui__NMi-Dp')
            nickname = nickname.text.strip() if nickname else ''

            # ë¦¬ë·° ë³¸ë¬¸
            content = r.select_one('.pui__vn15t2')
            content = content.text.strip() if content else ''

            # ìž‘ì„±ì¼ & ìž¬ë°©ë¬¸ ì—¬ë¶€
            dates = r.select('.pui__gfuUIT')
            date = dates[0].text.strip() if len(dates) > 0 else 'N/A'
            revisit = dates[1].text.strip() if len(dates) > 1 else 'N/A'

            # ì €ìž¥ ë° ì¶œë ¥
            print(nickname, '/', content, '/', date, '/', revisit)
            review_list.append([store_id, nickname, content, date, revisit])
            time.sleep(0.5)

        print(f"\nâœ… ì´ {len(review_list)}ê°œì˜ ë¦¬ë·° ìˆ˜ì§‘ ì™„ë£Œ")
        return store_dict, review_list

    except Exception as e:
        print(f"ì—ëŸ¬ ë°œìƒ: {e}")
        return store_dict, review_list
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import json

def scrape_store_info_with_reviews(store_id, max_page_downs=30, save_json=False):
    url = f'https://m.place.naver.com/restaurant/{store_id}/review/visitor?entry=plt&reviewSort=recent'
    
    store_dict = {"store_id": store_id, "status": "fail"}
    review_list = []
    
    # ë“œë¼ì´ë²„ ì˜µì…˜ ì„¤ì • (ì°½ ìˆ¨ê¹€)
    options = Options()
    #options.add_argument("--headless")  # ì°½ ìˆ¨ê¹€
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    driver = None
    try:
        driver = webdriver.Chrome(options=options)
        driver.get(url)
        driver.implicitly_wait(10)

        # â”€â”€ ê°€ê²Œ ì •ë³´ ìˆ˜ì§‘ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        time.sleep(2)
        try:
            store_name = driver.find_element(By.XPATH, '//*[@id="_title"]/div/span[1]').text
            store_dict["store_name"] = store_name
        except:
            store_dict["store_name"] = "ì •ë³´ ì—†ìŒ"

        try:
            store_dict["participation_count"] = driver.find_element(By.CSS_SELECTOR, 'div.jypaX > em').text
            store_dict["participation"] = driver.find_element(By.CSS_SELECTOR, 'div.jypaX > span.mholG').text
        except:
            store_dict["participation_count"] = "ì •ë³´ ì—†ìŒ"
            store_dict["participation"] = "ì •ë³´ ì—†ìŒ"

        try:
            tags = driver.find_elements(By.CSS_SELECTOR, 'div.vfTO3 > span.t3JSf')
            store_dict["top3_tags"] = [tag.text for tag in tags[:3]]
        except:
            store_dict["top3_tags"] = []

        # â”€â”€ ë¦¬ë·° ìˆ˜ì§‘ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        for _ in range(max_page_downs):
            try:
                more_button = driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[6]/div[2]/div[3]/div[2]/div/a')
                more_button.click()
                time.sleep(1.2)
            except:
                break

        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, 'lxml')
        reviews = soup.select('li.place_apply_pui.EjjAW')

        for r in reviews:
            nickname = r.select_one('.pui__NMi-Dp')
            content = r.select_one('.pui__vn15t2')
            dates = r.select('.pui__gfuUIT')

            review_data = {
                "nickname": nickname.text.strip() if nickname else '',
                "content": content.text.strip() if content else '',
                "date": dates[0].text.strip() if len(dates) > 0 else 'N/A',
                "revisit": dates[1].text.strip() if len(dates) > 1 else 'N/A'
            }
            review_list.append(review_data)

        store_dict["reviews"] = review_list
        store_dict["review_count"] = len(review_list)
        store_dict["status"] = "success"

        print(f"âœ… ìˆ˜ì§‘ ì™„ë£Œ: {store_dict['store_name']} - {store_dict['review_count']}ê°œ ë¦¬ë·°")

        # ì˜µì…˜: JSONìœ¼ë¡œ ì €ìž¥
        if save_json:
            with open(f"{store_id}_reviews.json", "w", encoding="utf-8") as f:
                json.dump(store_dict, f, ensure_ascii=False, indent=2)
            print("ðŸ“ JSON íŒŒì¼ ì €ìž¥ ì™„ë£Œ!")

    except Exception as e:
        print(f"âŒ ì—ëŸ¬ ë°œìƒ: {e}")
        store_dict["error"] = str(e)

    finally:
        if driver:
            driver.quit()

    return store_dict
```
